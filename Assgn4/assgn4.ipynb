```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Fashion_Retail_Sales.csv")

df['Date Purchase'] = pd.to_datetime(df['Date Purchase'], format="%d-%m-%Y", errors='coerce')

df_clean = df.dropna(subset=['Purchase Amount (USD)', 'Date Purchase'])
daily_sales = df_clean.groupby('Date Purchase')['Purchase Amount (USD)'].sum().reset_index()

plt.figure(figsize=(12,6))
plt.plot(daily_sales['Date Purchase'], daily_sales['Purchase Amount (USD)'], marker='o', linestyle='-')
plt.title('Time Series of Total Purchase Amount (USD)')
plt.xlabel('Date')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
